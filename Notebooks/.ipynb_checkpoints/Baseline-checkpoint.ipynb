# Load data

```python
# Load dataset
training = Dataset('../Data/NYU_Silberman_100.mat')
testing = Dataset('../Data/NYU_Silberman_101_120.mat')
```

# Loss function

```python
def mean_dot_product(y_true, y_pred):
    dot = tf.einsum('ijkl,ijkl->ijk',tf.nn.l2_normalize(y_true,3),tf.nn.l2_normalize(y_pred,3))
    n = tf.cast(tf.count_nonzero(dot),tf.float32)
    mean = tf.reduce_sum(dot) / n
    return -1 * mean
```

```python
norms[1,23,50,0]
```

```python
x = norms
y = np.copy(x)
y[1,23,50,0] = -0.87
tfx = tf.constant(x)
tfy = tf.constant(y)
```

```python
loss = mean_dot_product(tfx,tfy)

sess = tf.Session()
np.array(sess.run(loss))
```

# Baseline Model

```python
def baseline_model():
    # create model
    model = Sequential()
    model.add(Conv2D(12, (5, 5), input_shape=(240, 320, 3), activation='relu', padding='same'))
    model.add(Conv2D(32, (5, 5), activation='relu', padding='same'))
    model.add(Conv2D(12, (5, 5), activation='relu', padding='same'))
    model.add(Conv2D(3, (5, 5), activation='relu', padding='same'))
    # Compile model
    model.compile(loss= mean_dot_product, optimizer='sgd')
    return model
```

# Training

```python
model = baseline_model()
```

```python
for epoch in range(20):
    print('My Epoch', epoch)
    imgs , norms , masks = training.get_batch(100, 320, 240)
    x = imgs
    y = np.multiply(masks,norms)
    model.fit(x, y, batch_size=1, epochs=3)
```

# Prediction
